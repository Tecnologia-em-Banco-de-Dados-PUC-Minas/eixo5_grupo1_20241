# Pré-processamento dos dados

### Padronizando os títulos das colunas para garantir consistência no DataFrame

```python
# Renomeia as colunas da base df_2023 retirando o "_%"
for column in df_2023.columns:
  if "_%" in column:
    df_2023 = df_2023.rename(columns={column: column.replace("_%", "")})


# Renomeia a coluna "released_year" da base df_2023 para "year"
df_2023 = df_2023.rename(columns={"released_year": "year"})


# Renomeia a coluna "track_popularity" da base df_best_songs para "popularity"
df_best_songs = df_best_songs.rename(columns={"track_popularity": "popularity"})

# Renomeia a coluna "title" da base df_best_songs para "track_name"
df_best_songs = df_best_songs.rename(columns={"title": "track_name"})


# Deixa todos os títulos das músicas minusculos
df_2023['track_name'] = df_2023['track_name'].str.lower()
df_best_songs['track_name'] = df_best_songs['track_name'].str.lower()
df_spotify_songs['track_name'] = df_spotify_songs['track_name'].str.lower()
df_unpopular['track_name'] = df_unpopular['track_name'].str.lower()
df_2010to2022['track_name'] = df_2010to2022['track_name'].str.lower()
```

### Verifica colunas comuns a todos os dataset após as alterações

```python
# Armazena os nomes das colunas presentes em cada um dos dataset
columns_2023 = df_2023.columns
columns_2010_2022 = df_2010to2022.columns
columns_unpopular = df_unpopular.columns

# Imprime as colunas comuns a todos os dataset
column_commom = columns_unpopular.intersection(columns_2023).intersection(columns_2010_2022)
print(column_commom, column_commom.size)
```

```python
df_best_songs = df_best_songs.dropna()
df_2023 = df_2023.dropna()
df_2010to2022 = df_2010to2022.dropna()
df_unpopular = df_unpopular.dropna()
df_spotify_songs = df_spotify_songs.dropna()
```
